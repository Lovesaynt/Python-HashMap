```python
class Hashmaps:
# =================================================== #

    def __init__(self, size=100):
        self.size = size
        self.buckets  = [[] for _ in range(size)]

# =================================================== #

    def _hash(self, key):
        return hash(key) % self.size
    
# =================================================== #

    def put(self, key, value):
        index = self._hash(key)
        bucket = self.buckets[index]

        for i, (k, v) in enumerate (bucket):
            if k == key:
                bucket[i] = (key, value)
        bucket.append((key,value))

# =================================================== #

    def get(self, key):
        index = self._hash(key)
        bucket = self.buckets[index]

        for k, v in bucket:
            if k == key:
                return v
        raise KeyError('key')
    
# =================================================== #

    def remove(self, key):
        index = self._hash(key)
        bucket = self.buckets[index]

        for i, (k, _) in enumerate(bucket):
            if k == key:
                del bucket[i]
                return
        raise KeyError(key)
    
# =================================================== #

    def debug(self, key):
        index = self._hash(key)
        bucket = self.buckets[index]

        print('DEBUG ')
        print(f'Chave: {key}')
        print(f'√çndice: {index}')
        print(f'Bucket: {bucket}')

# =================================================== #

    def show(self):
        for i, bucket in enumerate(self.buckets):
            print(f'{i}: {bucket}')
            
# =================================================== #
```

Definindo nosso hashmap

```python
hm = Hashmaps(size = 10)

hm.put('Nome', 'Pablo')
hm.put('Idade', 24)
hm.put('Sexo', 'Masculino')
hm.put('Pais', 'Brasil')
hm.put('Estado', 'Vivo')
```

Debug

```python
hm.debug('Nome')
```

Exibir Buckets

```python
hm.show()
```
